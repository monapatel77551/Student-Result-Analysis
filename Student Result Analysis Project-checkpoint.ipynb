# Analysis OF the data--: GENDER DISTRIBUTION

```python
plt.figure(figsize=(5,5))
plt.title("GENDER DISTRIBUTION IN A SCHOOL")
ax=sns.countplot(data=df,x='Gender') # here we are using the countplot method of seaborn to check gender distribution
for cnt in ax.containers:
    ax.bar_label(cnt)
#ax.bar_label(ax.containers[0])
#plt.show()
```

From the above chart we can analyse that female are more than males..

```python
# here we are going toanalyse the impact of parents over the students group based on few factor
parent_education_impact=df.groupby('ParentEduc').agg({"MathScore":'mean',"ReadingScore":'mean','WritingScore':'mean'})
parent_education_impact
```

# Analysis of Impact of parents education on their children....

```python
sns.heatmap(parent_education_impact,annot=True)# we have used heatmap for better analysis
plt.title('Relationship b/w Parents Education  &  Score of Students')
plt.show()
```

From the above analysis we can find out that parents who are more educated their students are capable enough to score good marks

```python
marital_status=df.groupby('ParentMaritalStatus').agg({"MathScore":'mean',"ReadingScore":'mean','WritingScore':'mean'})
marital_status
```

```python
sns.heatmap(marital_status,annot=True) # we have used heatmap for better analysis
plt.title("Marital Status Impact On Students Score")
plt.show()
```

From above analysis we can find out that parents marital status have negligible impact on students score...
